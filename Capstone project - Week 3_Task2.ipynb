# Week 3 Task 2 Fetch Geografical Locations for Postal Codes

```python
#Import libraries
import numpy as np
import pandas as pd

#Read csv file exported in Task1
df = pd.read_csv("PostalCodes.csv")
```

## **Step1.**Download Coordinates and create new data frame

```python
#Download Location Data
!wget https://cocl.us/Geospatial_data
```

```python
#Create data frame with locations coordinates
loc = pd.read_csv('Geospatial_data')
loc.columns = ['Postcode', 'Latitude', 'Longitude']
loc.head()
```

## **Step 2.** Combine two data frames

```python
#Combine both data frames
post_loc = df.join(loc)
post_loc
```

```python
#Size of the data frame
post_loc.shape
```

```python
#Export csv file for future work
post_loc.to_csv('PostalCodes2.csv', index = False)
```
